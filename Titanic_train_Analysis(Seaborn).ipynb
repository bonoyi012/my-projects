```python
# Change to make subplots based on study time
sns.relplot(x='Sex', y='Age',
           data = df,
           kind = 'scatter',
           col = 'Pclass',
           hue = 'Survived')
plt.show()
```

```python
sns.relplot(data = df, x = 'Age', y = 'Fare', kind = 'line', style='Pclass', ci=None)
# ci => simply computes the percentile limites. The actual mean, the standard deviation, etc. will appear in the percentiles routine.
# ci = 'sd'
```

```python
# catplots (categorical values) - relational plots
```

```python
# create column subplots based on age catgory
sns.catplot(y='Survived', data=df, kind='count')
```

```python
# distinguish sex, male and female
sns.catplot(data=df,y='Survived', kind='count', hue='Sex')
```

```python
# add column (Pclass)
sns.catplot(data=df, y='Survived', kind='count', hue='Sex', col='Pclass')
```

```python
# create colum subplots based on agae category
sns.catplot(y='Survived', data=df, kind='bar', col='Pclass', order=['Pclass=1','Pclass=2','Pclass=3'], ci=None).fig.suptitle('Survival by class', y =1.2)
# about suptitle: it's indicating 'super'title. It appears at the top of the figure. The difference with normal title is when you multiple subplots on the same figure and you would like a title at the top of the figure not on each of the axes objects
```

```python
sns.catplot(data=df, sym="", x="Pclass")
```
